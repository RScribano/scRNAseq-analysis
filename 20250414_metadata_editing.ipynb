```python
import os
import pandas as pd
import numpy as np
import re
```

```python
original_metadata_path = r"C:\Users\ricca\Documents\cyTOF data\cohort 1\leukemia panel\20250414_scrnaseq metadata.csv"
original_metadata = pd.read_csv(original_metadata_path)
```

```python
# define function
def newcol(row):

    patient_name_old = row["individual"]

    if isinstance(patient_name_old, str):
        patient_name_new = re.sub("_", "", patient_name_old)
        timepoint = row["timepoint"]
        print(f"{patient_name_new}; {timepoint}")

        sample_name = f"{patient_name_new}_{timepoint}"

    else:
        print("invalid type")
        sample_name = np.nan

    return sample_name

# generate new column compatible with sce ColData
original_metadata["Sample_Name"] = original_metadata.apply(newcol, axis=1)
```

```python
# export new metadata 

sce_metadata =  original_metadata[original_metadata["condition"] == "basal"][["Sample_Name", "MRD_risk", "individual", "relapse", "timepoint"]]
sce_metadata
sce_metadata.to_csv(r"C:\Users\ricca\Documents\scRNAseq data\20250414_Metadata_scRNAseq.csv", index=None)

```
